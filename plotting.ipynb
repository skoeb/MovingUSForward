```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')

neighbors = pd.read_csv('assets/neighborhoods.csv')

stops = pd.read_csv('assets/police.csv')
stops = stops.rename({'Neighborhood':'Neighborhoods'}, axis='columns')

df = neighbors.merge(stops, on='Neighborhoods')

fig, ax = plt.subplots(figsize=(12,4))
df = df.set_index('Neighborhoods')
df.plot.bar(rot = 0, ax=ax)

ax.set_title("Top Neighborhoods with Highest Percent of Hispanic vs. White Population", size=16)
ax.set_ylabel('Percent')


plt.savefig('assets/neighborhood_percents.png')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

stops = pd.read_csv('assets/police.csv')
stops = stops.drop('Unnamed: 2', axis='columns')

fig, ax = plt.subplots(figsize=(10,5))
stops = stops.set_index('Neighborhood')
stops.plot.bar(rot = 0, ax=ax, legend=False)

ax.set_title("Police Stops by Neighborhood", size=16)
ax.set_ylabel('Percent')

plt.savefig('assets/police.png')
```
